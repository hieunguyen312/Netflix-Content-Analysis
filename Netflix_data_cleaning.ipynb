```python
# Check for possible null indicators
value_columns = {}

for column in df.columns:
    for value in df[column].unique():
        if value not in value_columns:
            value_columns[value] = []
        value_columns[value].append(column)

# Convert to df
value_columns_df = pd.DataFrame([(value, cols, len(cols)) for value, cols in value_columns.items()], 
                                columns=["Value", "Columns", "Column Count"])

# Filter values appearing in multiple columns
suspected_nulls = value_columns_df[value_columns_df["Column Count"] > 1]

print(suspected_nulls)
# "Not Given" is the null indicator
# Appear in director and country cols
```

```python
# Print the df when "Not Given" appeared in director
df[df["director"] == "Not Given"]
```

```python
# Print the df when "Not Given" appeared in contry
df[df["country"] == "Not Given"]
```

```python
# Count how much the null values apeared
director_null = (df["director"] == "Not Given").sum()
print(director_null)
print(str(round(director_null/len(df)*100,2)) + "%")
country_null = (df["country"] == "Not Given").sum()
print(country_null)
print(str(round(country_null/len(df)*100,2)) + "%")
# Since the null value for director is nearly 30%, we will leave it as "Not Given"
```

```python
# Drop all the duplicates
df.drop_duplicates()
```

```python
# Spliting the list_in col for future use using one hot encoder
# There appears to be non-string values in listed_in
# Convert every value to a string
# Every values in listed_in got turn into "nan:
# Inspecting data type
#print(df["listed_in"].dtype)
df["listed_in"] = df["listed_in"].apply(str)
df


```

```python
# Get a list of every genres
genres = df['listed_in'].str.split(', ').tolist()
#print(genres)
# Flatten list
flat_genres = [item for sublist in genres for item in sublist]
#print(flat_genres)
# Make a unique list
set_genres = set(flat_genres)
unique_genres = list(set_genres)
# No null values
#unique_genres.remove('NA')
print(unique_genres)
```

```python
df = df.reindex(df.columns.tolist() + unique_genres, axis=1, fill_value= 0 )

# for each value inside column, update the dummy
for index, row in df.iterrows():
    for val in row.listed_in.split(', '):
        if val != 'NA':
            df.loc[index, val] = 1
df.drop('listed_in', axis = 1, inplace = True)  
```
